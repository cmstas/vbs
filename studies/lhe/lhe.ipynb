```python
import uproot
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
lumi_inv_fb = 138

sm_xsec_pb = 7.46969E-02
with uproot.open("output/sm_vbswh.root") as f:
    sm_df = f.get("tree").arrays(library="pd")
    sm_df["xsec_weight"] = (1000*sm_xsec_pb*lumi_inv_fb)/10000
    
mkw_xsec_pb = 4.32674E-01
with uproot.open("output/mkw_vbswh.root") as f:
    mkw_df = f.get("tree").arrays(library="pd")
    mkw_df["xsec_weight"] = (1000*mkw_xsec_pb*lumi_inv_fb)/10000

mkz_xsec_pb = 4.32675E-01
with uproot.open("output/mkz_vbswh.root") as f:
    mkz_df = f.get("tree").arrays(library="pd")
    mkz_df["xsec_weight"] = (1000*mkz_xsec_pb*lumi_inv_fb)/10000
```

```python
mkw_mkz_concat_df = pd.concat(
    [mkw_df.drop(columns=["xsec_weight"]), mkz_df.drop(columns=["xsec_weight"])]
).reset_index(drop=True)

# Find duplicates
mkw_mkz_duplicate_df = mkw_mkz_concat_df[mkw_mkz_concat_df.duplicated(keep=False)].sort_values("H_pt")
# Label mkw or mkz
mkw_mkz_duplicate_df["sample"] = ""
mkw_mkz_duplicate_df.loc[mkw_mkz_duplicate_df.index < 10000, "sample"] = "mkw"
mkw_mkz_duplicate_df.loc[mkw_mkz_duplicate_df.index >= 10000, "sample"] = "mkz"
mkw_mkz_duplicate_df.to_csv("output/duplicate.csv", index=False)
mkw_mkz_duplicate_df
```

```python
mkw_mkz_concat_df = pd.concat(
    [mkw_df.drop(columns=["xsec_weight"]), mkz_df.drop(columns=["xsec_weight"])]
).reset_index(drop=True)

# Find duplicates
mkw_mkz_nodupl_df = mkw_mkz_concat_df[~np.isin(mkw_mkz_concat_df.index, mkw_mkz_duplicate_df.index)].sort_values("H_pt")
# Label mkw or mkz
mkw_mkz_nodupl_df["sample"] = ""
mkw_mkz_nodupl_df.loc[mkw_mkz_nodupl_df.index < 10000, "sample"] = "mkw"
mkw_mkz_nodupl_df.loc[mkw_mkz_nodupl_df.index >= 10000, "sample"] = "mkz"
mkw_mkz_nodupl_df.to_csv("output/unique.csv", index=False)
mkw_mkz_nodupl_df
```

```python
def smart_clip(np_array, bins):
    if len(bins) < 2:
        raise ValueError("must have at least two bins")
        return
    clip_low = 0.5 * (bins[0] + bins[1])
    clip_high = 0.5 * (bins[-2] + bins[-1])
    return np.clip(np_array, clip_low, clip_high)

def plot(var, bins=None, query="index >= 0", x_label="", transf=lambda x: x):
    fig, axes = plt.subplots(figsize=(12, 8))
    # Plot SM
    global sm_df
    sm_weights = sm_df.query(query).xsec_weight
    axes.hist(
        smart_clip(transf(sm_df.query(query)[var]), bins), 
        bins=bins,
        label=f"SM [{np.sum(sm_weights):0.1f} events]",
        alpha=0.5,
        weights=sm_weights
    );
    # Plot -kW
    global mkw_df
    mkw_weights = mkw_df.query(query).xsec_weight
    axes.hist(
        smart_clip(transf(mkw_df.query(query)[var]), bins), 
        bins=bins, 
        histtype="step",
        label=f"$-\kappa_W$ [{np.sum(mkw_weights):0.1f} events]",
        weights=mkw_weights
    );
    # Plot -kZ
    global mkz_df
    mkz_weights = mkz_df.query(query).xsec_weight
    axes.hist(
        smart_clip(transf(mkz_df.query(query)[var]), bins), 
        bins=bins, 
        histtype="step",
        label=f"$-\kappa_Z$ [{np.sum(mkz_weights):0.1f} events]",
        weights=mkz_weights
    );
    # Plot luminosity information
    global lumi_inv_fb
    axes.text(
        1, 1, 
        f"{lumi_inv_fb} fb$^{{-1}}$ (13 TeV)",
        horizontalalignment="right",
        verticalalignment="bottom",
        size=24,
        transform=axes.transAxes
    )
    # Plot legend
    axes.legend(fontsize=14);
    # Change axis labels
    axes.set_xlabel(x_label, size=24);
    axes.set_ylabel("Events", size=24);
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    return axes
```

```python
plot("H_pt", bins=np.linspace(0, 1000, 21), x_label="$p_{T,H}$ [GeV]");
plot("W_pt", bins=np.linspace(0, 1000, 21), x_label="$p_{T,W}$ [GeV]");
plot("M_WH", bins=np.linspace(0, 2500, 21), x_label="$M_{WH}$ [GeV]");
```

```python
plot("W_pt", bins=np.linspace(0, 1000, 21), query="H_pt > 400", x_label="$p_{T,W}$ [GeV]");
plot("M_WH", bins=np.linspace(0, 5000, 21), query="H_pt > 400", x_label="$M_{WH}$ [GeV]");
```

```python
axes = plot("W_sign", bins=[-1, 0, 1, 2], x_label="$sign(W)$", transf=np.sign)
axes.set_xticks([-0.5, 1.5]);
axes.set_xticklabels([-1, 1]);

axes = plot("W_pol", bins=[-1, 0, 1, 2], x_label="$polarization(W)$", transf=np.sign)
axes.set_xticks([-0.5, 0.5, 1.5]);
axes.set_xticklabels([-1, 0, 1]);
```
